```python
import pandas as pd
import statsmodels.api as sm
```

```python
wind=pd.read_csv('device_position.csv')
mac=pd.read_csv('mac_lat.csv')
temp=pd.read_csv('temp_hum.csv')
```

```python
temp
```

```python
newm=temp.merge(mac, on='mac')
newm
```

```python
wind
```

```python
total=newm.merge(wind, left_on=['day', 'hour', 'dev_lat', 'dev_long'], right_on=['day', 'hour', 'lat', 'lon'])
total
```

```python
X=total[['avg_temp', 'avg_hum']]
y=total[['avg_wind_speed']]
```

```python
X=sm.add_constant(X)
model=sm.OLS(y, X)
result=model.fit()
result.summary()
```
